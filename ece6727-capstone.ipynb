**Concat the Data**

```python
import pandas as pd

# Load the CSV files into DataFrames
drone_data_df = pd.read_csv('/kaggle/input/drone-based-malware-detection-dbmd/drone_data.csv')
environmental_data_df = pd.read_csv('/kaggle/input/drone-based-malware-detection-dbmd/environmental_data.csv')
malware_detection_df = pd.read_csv('/kaggle/input/drone-based-malware-detection-dbmd/malware_detection_data.csv')
network_traffic_df = pd.read_csv('/kaggle/input/drone-based-malware-detection-dbmd/network_traffic_data.csv')


# Ensure all datasets have the same number of rows
assert drone_data_df.shape[0] == environmental_data_df.shape[0] == malware_detection_df.shape[0] == network_traffic_df.shape[0], "The CSV files do not have the same number of rows."

# Concatenate the dataframes along the columns (axis=1)
merged_df = pd.concat([drone_data_df, environmental_data_df, malware_detection_df,network_traffic_df], axis=1)

# Display the first few rows of the merged dataset
print("Merged Dataset:")
display(merged_df.head())

# Display the shape of the merged dataset to verify
print("Shape of the merged dataset:", merged_df.shape)

# Save the merged dataset to a new CSV file
merged_df.to_csv('merged_dataset.csv', index=False)


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)  # Suppress the warning

# Univariate Analysis for Continuous Variables
continuous_columns = ['latitude', 'longitude', 'altitude', 'speed', 'battery_level', 
                      'signal_strength', 'temperature', 'humidity', 'wind_speed', 'pressure']

plt.figure(figsize=(15, 10))
for i, col in enumerate(continuous_columns):
    plt.subplot(4, 3, i + 1)
    sns.histplot(merged_df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Univariate Analysis for Categorical Variables
categorical_columns = ['weather_conditions', 'attack_type', 'location_type']

for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=merged_df)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bivariate Analysis: Correlation Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = merged_df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Pairplot (Optional): To see pairwise relationships
sns.pairplot(merged_df[continuous_columns])
plt.show()
```

```python
# Boxplot for Categorical Variables
plt.figure(figsize=(10, 6))
sns.boxplot(x='weather_conditions', y='speed', data=merged_df)
plt.title("Boxplot of Speed vs Weather Conditions")
plt.xticks(rotation=45)
plt.show()
```

```python
# Target Variable Analysis
print("Attack Type Distribution:")
print(merged_df['attack_type'].value_counts())

plt.figure(figsize=(10, 6))
sns.countplot(x='attack_type', data=merged_df)
plt.title("Distribution of Attack Types")
plt.xticks(rotation=45)
plt.show()
```

```python
# Outlier Detection using Boxplot for Numerical Features
plt.figure(figsize=(15, 10))
for i, col in enumerate(continuous_columns):
    plt.subplot(4, 3, i + 1)
    sns.boxplot(x=merged_df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np

# Advanced Outlier Detection using Z-score
z_scores = np.abs(stats.zscore(merged_df[continuous_columns]))
outliers = (z_scores > 3).sum(axis=1)  # Detect outliers with z-score > 3
print(f"Number of outliers: {np.sum(outliers)}")
```

```python
from sklearn.decomposition import PCA

# Apply PCA on the continuous features
numeric_df = merged_df[continuous_columns]
pca = PCA(n_components=2)
pca_result = pca.fit_transform(numeric_df)
print(f'Explained variance by each component: {pca.explained_variance_ratio_}')
```
